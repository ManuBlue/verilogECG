```python
import pandas as pd

# Load the datasets
df1 = pd.read_csv('Data/mitbih_train.csv')
df2 = pd.read_csv('Data/mitbih_test.csv')

# Multiply all columns except the last one by 4096 and round to the nearest integer
df1.iloc[:, :-1] = (df1.iloc[:, :-1] * 4096).round().astype(int)
df2.iloc[:, :-1] = (df2.iloc[:, :-1] * 4096).round().astype(int)

# Rename columns
num_features = df1.shape[1] - 1  # Excluding last column
new_column_names = ['val' + str(i+1) for i in range(num_features)] + ['class']
df1.columns = new_column_names
df2.columns = new_column_names

# Ensure all values, including class, are integers
df1 = df1.astype(int)
df2 = df2.astype(int)
```

```python
# Save the modified DataFrames
df1.to_csv('Data/mitbih_train_processed.csv', index=False)
df2.to_csv('Data/mitbih_test_processed.csv', index=False)

print("Processing complete. Files saved.")
```
