```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("C:/Users/Lenovo/Downloads/mitbih_test.csv", header=None)

def process_ecg_signal(ecg_signal):
    sorted_indices = np.argsort(ecg_signal)[-2:]
    start_index, end_index = sorted_indices.min(), sorted_indices.max()

    processed_signal = np.zeros_like(ecg_signal)
    processed_signal[start_index:end_index+1] = ecg_signal[start_index:end_index+1]

    return processed_signal

processed_signals = data.apply(process_ecg_signal, axis=1)

plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.plot(data.iloc[0], label='Original ECG Signal')
plt.legend()
plt.subplot(2, 1, 2)
plt.plot(processed_signals.iloc[0], label='Processed ECG Signal')
plt.legend()
plt.show()
```
